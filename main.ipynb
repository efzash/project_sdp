```python
import pandas_datareader as pdr
import pandas as pd
import datetime 
```

```python
tatn = pdr.get_data_yahoo('TATN.ME', 
                          start=datetime.datetime(2015, 1, 1), 
                          end=datetime.datetime(2020, 5, 1))
lukoil = pdr.get_data_yahoo('LKOH.ME', 
                          start=datetime.datetime(2015, 1, 1), 
                          end=datetime.datetime(2020, 5, 1))
bashn = pdr.get_data_yahoo('BANEP.ME', 
                          start=datetime.datetime(2015, 1, 1), 
                          end=datetime.datetime(2019, 12, 1))
```

```python
# fig = plt.line(tatn, tatn.index, y="Close")
# fig.show()

from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=("Tatneft", "Lukoil", "Bashneft"))

fig.add_trace(go.Scatter(x=tatn.index, y=tatn["Close"], name="Tatneft"),
              row=1, col=1)

fig.add_trace(go.Scatter(x=lukoil.index, y=lukoil["Close"],name="Lukoil"),
              row=2, col=1)

fig.add_trace(go.Scatter(x=bashn.index, y=bashn["Close"], name="Bashneft"),
              row=3, col=1)

fig.update_layout(height=500, width=700,
                  title_text=" Subplots with Titles")

fig.show()
```

```python
# Import `numpy` as `np`
import numpy as np

# Assign `Adj Close` to `daily_close`
tatn_daily_close = tatn[['Adj Close']]
lukoil_daily_close = lukoil[['Adj Close']]
bashn_daily_close = bashn[['Adj Close']]
# Assign 'Volume' to 'daily_volume'
tatn_daily_volume = tatn[['Volume']]
lukoil_daily_volume = lukoil[['Volume']]
bashn_daily_volume = bashn[['Volume']]
# 
tatn_daily_volume_pct_change = tatn_daily_volume.pct_change()
lukoil_daily_volume_pct_change = lukoil_daily_volume.pct_change()
bashn_daily_volume_pct_change = bashn_daily_volume.pct_change()
# Daily returns
tatn_daily_pct_change = tatn_daily_close.pct_change()
lukoil_daily_pct_change = lukoil_daily_close.pct_change()
bashn_daily_pct_change = bashn_daily_close.pct_change()

# Replace NA values with 0
tatn_daily_pct_change.fillna(0, inplace=True)
lukoil_daily_pct_change.fillna(0, inplace=True)
bashn_daily_pct_change.fillna(0, inplace=True)
#
tatn_daily_volume_pct_change.fillna(0, inplace=True)
lukoil_daily_volume_pct_change.fillna(0, inplace=True)
bashn_daily_volume_pct_change.fillna(0, inplace=True)
# Inspect daily returns
print(tatn_daily_pct_change)
print(tatn_daily_volume_pct_change)
```

```python
# fig = plt.line(daily_pct_change, daily_pct_change.index, y="Adj Close")
# fig.show()


fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=("Tatneft", "Lukoil", "Bashneft"))

fig.add_trace(go.Scatter(x=tatn_daily_pct_change.index, y=tatn_daily_pct_change["Adj Close"], name="Tatneft"),
              row=1, col=1)

fig.add_trace(go.Scatter(x=lukoil_daily_pct_change.index, y=lukoil_daily_pct_change["Adj Close"],name="Lukoil"),
              row=2, col=1)

fig.add_trace(go.Scatter(x=bashn_daily_pct_change.index, y=bashn_daily_pct_change["Adj Close"], name="Bashneft"),
              row=3, col=1)

fig.update_layout(height=500, width=700,
                  title_text="Daily stock value changes in percent")

fig.show()
```
